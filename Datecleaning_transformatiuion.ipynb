```python
# part 1: cleaning and transforming the check_in_date column

#cleaning part
#standardize the ddelimiters-like use hypens in all the dates and replace or remove any slashes
df_ke_hotels['check_in_date_cleaned']=df_ke_hotels['check_in_date'].str.replace("/", "-", regex=True)
```

```python
#replace the check in. date from string (object) to date time
df_ke_hotels['check_in_date_cleaned']=pd.to_datetime(df_ke_hotels['check_in_date_cleaned'],
 errors='coerce',
format ='mixed', # date formnat is mixed if yes follow up wqith dayfirst =true, otherwise if noit mixed omit the format and dayfirst
                                                     # during inspection, I noticed that the check_in_date column has multiple formats.
                      #So I will use format = 'mixed' to handle this issue.
                      # If you are sure that all the dates in the column are in the same format,
                      # you can specify that format instead of using 'mixed'
                      # examples: %Y-%m-%d, %d-%m-%Y, %m/%d/%Y,
 dayfirst=True) # during inspection when you see format as %Y-%m-%d, no need to use dayfirst = True
df_ke_hotels['check_in_date_cleaned'].head(10) # all date format are standardized in y/m/d and rest of errors coerced into NAT
```

```python
df_ke_hotels.info() # check_in_date_cleaned column has been added
```

```python
# the transformation part

# find check in year
df_ke_hotels['check_in _year']=df_ke_hotels['check_in_date_cleaned'].dt.year

# find check in month id and check in month name
df_ke_hotels['check_in _month_id']=df_ke_hotels['check_in_date_cleaned'].dt.month

df_ke_hotels['check_in _month_name']=df_ke_hotels['check_in_date_cleaned'].dt.month_name()

## find check in quarter
df_ke_hotels['check_in _quarter']=df_ke_hotels['check_in_date_cleaned'].dt.quarter
# find check in quarter
# find check in day of the month, day name, day of the week (monday=o,1,2 etc)
df_ke_hotels['check_in_day_of month']=df_ke_hotels['check_in_date_cleaned'].dt.day
df_ke_hotels['check_in_day_name']=df_ke_hotels['check_in_date_cleaned'].dt.day_name()
df_ke_hotels['check_in_day_of_week']=df_ke_hotels['check_in_date_cleaned'].dt.dayofweek
#
```

```python
# using day of week we can create a column "is_weekend" to identify whether it's true a day is on weekend or false
df_ke_hotels['check_in_is_weekend'] =df_ke_hotels['check_in_day_of_week'].isin([5,6])
df_ke_hotels.head (15)
```

```python
#Assignment ## part 1: cleaning and transforming the bookingtimestamp column
#inspection and understanding the data
df_ke_hotels['booking_timestamp'].unique()
df_ke_hotels.info()
df_ke_hotels['booking_timestamp'].head(20)
#cleaning part
#standardize the ddelimiters-like use hypens in all the dates and replace or remove any slashes
```

```python

df_ke_hotels['booking_timestamp'].head(10) #inspecting the booking stamp column --has mixed format, dashes etc

#standardize-standardize the  date delimiters-like use hypens in all the dates and replace or remove any slashes
df_ke_hotels['booking_timestamp_cleaned']= df_ke_hotels['booking_timestamp'].str.replace ("/", "-",regex=True)
#df_ke_hotels['booking_timestamp_cleaned'].head()
```

```python
#replace the bookingstamp column from string (object) to date time
df_ke_hotels['booking_timestamp_cleaned']= df_ke_hotels['booking_timestamp'].str.replace ("/", "-", regex=True)
#df_ke_hotels['booking_timestamp_cleaned'].head()
df_ke_hotels['booking_timestamp_cleaned']=pd.to_datetime (df_ke_hotels['booking_timestamp_cleaned'],
  errors='coerce',
  format='mixed',  #date format is mixed so, yes follow up with dayfirst =true
  dayfirst= True) ## during inspection the format arfe not only this  %Y-%m-%d but are m ixed , \thus i  need to use dayfirst = True

df_ke_hotels[['booking_timestamp', 'booking_timestamp_cleaned']].head(8)
```

```python
# the transformation part

# find check in year
df_ke_hotels['booking_timestamp_cleaned_year'] = df_ke_hotels['booking_timestamp_cleaned'].dt.year
#month id
df_ke_hotels['booking_timestamp_cleaned_month_id'] = df_ke_hotels['booking_timestamp_cleaned'].dt.month
#month name
df_ke_hotels['booking_timestamp_cleaned_month_name'] = df_ke_hotels['booking_timestamp_cleaned'].dt.month_name()
#quarter
df_ke_hotels['booking_timestamp_cleaned_quarter'] = df_ke_hotels['booking_timestamp_cleaned'].dt.quarter

# find check in day of the month, day name, day of the week
#find bookings_timesgtamp of the month
df_ke_hotels['bookings_timestamp_day_of_month'] = df_ke_hotels['booking_timestamp_cleaned'].dt.day
df_ke_hotels['bookings_timestamp_day__name'] = df_ke_hotels['booking_timestamp_cleaned'].dt.day_name()
#day of the week
df_ke_hotels['bookings_timestamp_day__of_week'] = df_ke_hotels['booking_timestamp_cleaned'].dt.dayofweek
#is weekend
df_ke_hotels['booking_timestamp_cleaned_is_weekend'] = df_ke_hotels['bookings_timestamp_day__of_week'].isin ([5,6])

df_ke_hotels.head(15)
```
